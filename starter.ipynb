````python
def save_images(imges_df,images_data,path_to_save_img):
    rows=imges_df["row"]
    img_names=imges_df["filenames"]
    for r,n in zip(rows,img_names):
        
    #Grab the image data in row-major form

        img = images_data[b'data'][r]

        # Consolidated stacking/reshaping from earlier
        target = np.dstack((
                        img[0:1024].reshape(32,32,1),
                        img[1024:2048].reshape(32,32,1),
                        img[2048:].reshape(32,32,1)
                         ))
        # Save the image
        try:

            save_img_path = os.path.join(path_to_save_img, n)

            plt.imsave(save_img_path, target)
            raw_save_img_path = r"{}".format(save_img_path)
            nraw_save_img_path = os.path.normpath(raw_save_img_path)
            print(f"Successfully saved {n} to folder {nraw_save_img_path}.")

        except RuntimeError as e:
            return f"Error: {e} \nAn error accures while saving {n} file to the target folder located at {path_to_save_img}"
````

## Model Training

For Image Classification, Sagemaker [also expects metadata](https://docs.aws.amazon.com/sagemaker/latest/dg/image-classification.html) e.g. in the form of TSV files with labels and filepaths. We can generate these using our Pandas DataFrames from earlier:

````python
def to_metadata_file(df, prefix):
    df["s3_path"] = df["filenames"]
    df["labels"] = df["labels"].apply(lambda x: 0 if x==8 else 1)
    return df[["row", "labels", "s3_path"]].to_csv(
        f"{prefix}.lst", sep="\t", index=False, header=False
    )
    
to_metadata_file(df_train.copy(), "train")
to_metadata_file(df_test.copy(), "test")
````

````python
import random
import os
import json
endpoint_name = "image-classification-2025-08-29-15-32-08-957" 
class_names = ['bicycle', 'motorcycle']
def predict_on_image(image_path, predictor_instance, class_labels):
    """
    Takes an image path, sends it to the SageMaker endpoint for prediction,
    and returns the image and the prediction results.
    """
    if not predictor_instance:
        print("Predictor is not initialized. Cannot make a prediction.")
        return None, None

    try:
        # Read the image file in binary mode
        with open(image_path, "rb") as f:
            payload = f.read()

        # Get an inference from the endpoint
        inference = predictor_instance.predict(payload)

        # The inference is returned as bytes, so we need to decode it.
        # It's a JSON string representing a list of probabilities.
        probabilities = json.loads(inference.decode('utf-8'))
        
        # Find the predicted class index with the highest probability
        predicted_class_index = np.argmax(probabilities)
        
        # Get the corresponding class name and confidence score
        predicted_class_name = class_labels[predicted_class_index]
        confidence = probabilities[predicted_class_index]

        # Load the image for display
        img = plt.imread(image_path)
        
        return img, {
            "predicted_class": predicted_class_name,
            "confidence": f"{confidence:.2%}",
            "probabilities": {label: f"{prob:.2%}" for label, prob in zip(class_labels, probabilities)}
        }

    except FileNotFoundError:
        print(f"Error: The file '{image_path}' was not found.")
        return None, None
    except Exception as e:
        print(f"An error occurred during prediction for '{image_path}': {e}")
        return None, None

# --- Run Predictions on Sample Images ---
if predictor:
    # Get a list of all files in the ./test directory
    test_image_dir = './test/'
    try:
        test_images = [os.path.join(test_image_dir, f) for f in os.listdir(test_image_dir) if f.endswith('.png')]
        
        if not test_images:
            print(f"No PNG images found in the '{test_image_dir}' directory.")
        else:
            # Randomly select a few images to test
            num_samples = 5
            sample_images = random.sample(test_images, min(num_samples, len(test_images)))
            
            print(f"--- Making Predictions on {len(sample_images)} Sample Images ---\n")

            # Loop through the sample images and display predictions
            for image_file in sample_images:
                image, result = predict_on_image(image_file, predictor, class_names)
                
                if image is not None and result is not None:
                    plt.imshow(image)
                    plt.title(f"File: {os.path.basename(image_file)}\nPredicted: {result['predicted_class']} ({result['confidence']})")
                    plt.axis('off')
                    plt.show()
                    
                    print(f"File: {os.path.basename(image_file)}")
                    print(f"  - Predicted Class: {result['predicted_class']}")
                    print(f"  - Confidence: {result['confidence']}")
                    print(f"  - All Probabilities: {result['probabilities']}")
                    print("-" * 30)

    except FileNotFoundError:
        print(f"Error: The directory '{test_image_dir}' was not found. Make sure you have run the data staging part of the notebook.")
````

````python
import json
import os
import jsonlines
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Load and Parse the Captured Data ---
# This list will hold our parsed JSON objects
json_data = []

# Check if the directory exists and has files
if os.path.exists("./captured_data") and os.listdir("./captured_data"):
    file_handles = os.listdir("./captured_data")
    for jsonl_file in file_handles:
        full_path = os.path.join("./captured_data", jsonl_file)
        # Ensure we're only reading files, not directories like .ipynb_checkpoints
        if os.path.isfile(full_path):
            with jsonlines.open(full_path) as f:
                for line in f:
                    json_data.append(line)
else:
    print("Warning: 'captured_data' directory is empty. Plots will be empty.")
    
# --- 2. Extract Data for Plotting ---
# Define a helper function to get data from each JSON object
def simple_getter(obj):
    inferences = obj["captureData"]["endpointOutput"]["data"]
    timestamp = obj["eventMetadata"]["inferenceTime"]
    return json.loads(inferences), timestamp

# Create lists to hold the data for our axes
x_timestamps = []
y_confidence = []

for obj in json_data:
    inference, timestamp = simple_getter(obj)
    
    # The confidence is the highest probability in the inference list
    y_confidence.append(max(inference))
    # Append the timestamp string for the x-axis
    x_timestamps.append(timestamp)

# Convert timestamp strings to datetime objects for proper plotting
# This allows matplotlib to handle sorting and formatting automatically
if x_timestamps:
    x_timestamps = pd.to_datetime(x_timestamps)

# Sort the data by timestamp to ensure the plots are in chronological order
if not y_confidence.empty:
    sorted_data = sorted(zip(x_timestamps, y_confidence))
    x_timestamps, y_confidence = zip(*sorted_data)
print(f"Prepared data points for plotting.")
````

````python
# --- Create the Histogram ---
# The histogram only needs the confidence values (y_confidence)
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 6))

if y_confidence:
    plt.hist(y_confidence, bins=20, color='darkcyan', edgecolor='white', alpha=0.9)
    plt.axvline(x=0.93, color='red', linestyle='--', label='Production Threshold (0.93)')
else:
    plt.text(0.5, 0.5, 'No data to display', ha='center', va='center')


# --- Add Labels and Title ---
plt.xlabel("Inference Confidence", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.suptitle("Histogram of Observed Recent Inferences", size=16)
plt.legend()

plt.show()
````

````python
# --- Create the Line Plot ---
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(15, 7))

if x_timestamps:
    plt.plot(x_timestamps, y_confidence, lw=2.5, color='violet', alpha=1, marker='o', markersize=8)
    # This function automatically rotates the date labels for readability
    plt.gcf().autofmt_xdate()
else:
    plt.text(0.5, 0.5, 'No data to display', ha='center', va='center')

# --- Add Labels and Title ---
plt.suptitle("Observed Recent Inferences", size=16)
plt.xlabel("Time of Inference", fontsize=12)
plt.ylabel("Confidence", fontsize=12)
plt.ylim(bottom=min(y_confidence) - 0.1 if y_confidence else 0, top=1.05)


plt.show()
````

````python
# --- Create the Scatter Plot ---
plt.style.use("Solarize_Light2")
plt.figure(figsize=(15, 7))

if x_timestamps:
    # Plot the confidence scores against their actual timestamps
    plt.scatter(x_timestamps, y_confidence, c=['r' if k < 0.94 else 'b' for k in y_confidence], s=100)
    plt.axhline(y=0.94, color='g', linestyle='--')
    plt.ylim(bottom=0.60)
    # Automatically format the timestamp labels
    plt.gcf().autofmt_xdate()
else:
    plt.text(0.5, 0.5, 'No data to display', ha='center', va='center')


# --- Add Labels ---
plt.ylabel("Confidence")
plt.xlabel("Time of Inference")
plt.suptitle("Observed Recent Inferences", size=14)
plt.title("Pictured with confidence threshold for production use", size=10)


plt.show()
````
